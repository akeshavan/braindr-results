```python
for i, name in enumerate(df.indi_name):
    for si in sites:
        url = mb_tmpl.format(site=si, sub=name)
        if exists(url):
            df.loc[df['indi_name']==name, 'mb_url'] = url
            foo = pd.read_csv(url)
            foo.loc[foo['name'].isin(gm_names), 'type'] = "gray_matter"
            foo.loc[foo['name'].isin(wm_names), 'type'] = "white_matter"
            foo.loc[foo['name'].isin(csf_names), 'type'] = "csf"
            volumes = foo.groupby('type').apply(lambda x: x.sum()[' volume']).to_dict()
            for key, val in volumes.items():
                df.loc[df['indi_name']==name, key] = val
            break
```

```python
not_there = df[df.mb_url.isnull()].indi_name.values
```

```python
still_unknown = []
for n in not_there:
    url_tmpl = "http://s3.amazonaws.com/fcp-indi/data/Projects/HBN/MRI/{site}/derivatives/{sub}/freesurfer/scripts/recon-all.log"
    know = False
    for s in sites:
        url = url_tmpl.format(site=s, sub=n)
        if exists(url):
            print(n, "freesurfer error")
            df.loc[df['indi_name']==n, 'reason'] = "fs_error"
            know = True
            break
    if not know:
        print(n, "unknown")
        df.loc[df['indi_name']==n, 'reason'] = "unknown"
        still_unknown.append(n)
            
```
